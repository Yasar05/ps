```python
'''
In this assignment students have to compress racoon grey scale image into
5 clusters. In the end, visualize both raw and compressed image and look
for quality difference.
The raw image is available in spicy.misc package with the name face.
'''

# Importing libraries
import numpy as np
from sklearn.cluster import KMeans
import scipy.misc
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Visualization the gray scale image

g = scipy.misc.face(gray=True)
plt.figure(figsize=(10, 3.6))
plt.imshow(g, cmap=plt.cm.gray)
plt.show()
```

```python
# Compress the gray scale image into 5 clusters

rows = g.shape[0]
cols = g.shape[1]

#print(rows,cols)

image = g.reshape(rows*cols,1)
kmeans = KMeans(n_clusters = 5)
kmeans.fit(image)
```

```python
clusters = np.asarray(kmeans.cluster_centers_) 
labels = np.asarray(kmeans.labels_)  
labels = labels.reshape(rows,cols);
```

```python
plt.imsave('compressed_racoon.png',labels);
```

```python
# Visualization compressed image

img = plt.imread('compressed_racoon.png')
plt.figure(figsize=(10, 3.6))
plt.imshow(img)
plt.show()
```
